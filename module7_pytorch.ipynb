## Leçon 1 — Tenseurs et DataLoader

**PyTorch** est la bibliothèque de deep learning. Les **tenseurs** sont comme des tableaux NumPy mais permettent
le calcul de gradients automatique.

| Concept | Code |
|---|---|
| Array NumPy → Tenseur | `torch.from_numpy(arr)` |
| Dataset PyTorch | `TensorDataset(X_tensor, y_tensor)` |
| Chargeur de données | `DataLoader(dataset, batch_size=256, shuffle=True)` |

### Exemple

````python
import torch
import numpy as np
from torch.utils.data import TensorDataset, DataLoader
from sklearn.model_selection import train_test_split

# Données numpy → tenseurs PyTorch
X = np.random.randn(500, 9).astype(np.float32)
y = np.random.randn(500, 1).astype(np.float32)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Convertir en tenseurs
X_train_t = torch.from_numpy(X_train)
y_train_t = torch.from_numpy(y_train)
X_test_t  = torch.from_numpy(X_test)
y_test_t  = torch.from_numpy(y_test)

# Créer les datasets et DataLoaders
batch_size = 256
train_dataset = TensorDataset(X_train_t, y_train_t)
test_dataset  = TensorDataset(X_test_t,  y_test_t)

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_loader  = DataLoader(test_dataset,  batch_size=batch_size)

print(f"Batches par epoch: {len(train_loader)}")

# Itérer sur un batch
for xb, yb in train_loader:
    print(f"Batch shape: X={xb.shape}, y={yb.shape}")
    break
````

## Leçon 2 — Définir un Modèle (nn.Module)

En PyTorch, un modèle est une **classe** qui hérite de `nn.Module`. Elle a deux méthodes :
- `__init__` : définit les couches (ex : `self.lin = nn.Linear(in, out)`)
- `forward(x)` : décrit comment les données traversent le réseau

`nn.Linear(in_features, out_features)` = couche linéaire (régression linéaire). C'est exactement ce qui est utilisé
dans l'Exercice 2 de l'examen.

### Exemple

````python
import torch
import torch.nn as nn

# LinearReplicator: exactement le modèle de l'Exercice 2
class LinearReplicator(nn.Module):
    def __init__(self, in_features=9):
        super().__init__()
        # Une seule couche linéaire: in_features → 1 sortie
        self.lin = nn.Linear(in_features, 1, bias=True)

    def forward(self, x):
        # x passe à travers la couche linéaire
        return self.lin(x)

# Instancier le modèle
model = LinearReplicator(in_features=9)

# Voir les paramètres
for name, param in model.named_parameters():
    print(f"{name}: shape={param.shape}")

# Test forward pass
x_test = torch.randn(5, 9)   # batch de 5 observations, 9 features
out = model(x_test)
print(f"\nInput shape: {x_test.shape}")
print(f"Output shape: {out.shape}")   # (5, 1)
````

## Leçon 3 — La Boucle d'Entraînement

La boucle d'entraînement PyTorch a **5 étapes** répétées à chaque batch :
1. `optimizer.zero_grad()` — remet les gradients à zéro
2. `y_pred = model(xb)` — passe avant (forward pass)
3. `loss = criterion(y_pred, yb)` — calcule la perte
4. `loss.backward()` — calcule les gradients (rétropropagation)
5. `optimizer.step()` — met à jour les paramètres

⚠️ **Erreur fréquente :** oublier `optimizer.zero_grad()` fait s'accumuler les gradients des batchs précédents.

### Exemple

````python
import torch, torch.nn as nn
from torch.utils.data import TensorDataset, DataLoader

# Données
X = torch.randn(400, 9)
y = torch.randn(400, 1)
loader = DataLoader(TensorDataset(X, y), batch_size=64, shuffle=True)

# Modèle, loss, optimizer
model = nn.Linear(9, 1)
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)

n_epochs = 50

for epoch in range(1, n_epochs + 1):
    model.train()
    epoch_loss = 0.0

    for xb, yb in loader:
        # 5 étapes obligatoires
        optimizer.zero_grad()          # 1. Reset gradients
        y_pred = model(xb)              # 2. Forward pass
        loss = criterion(y_pred, yb)    # 3. Compute loss
        loss.backward()                 # 4. Backprop
        optimizer.step()                # 5. Update weights
        epoch_loss += loss.item() * xb.size(0)

    epoch_loss /= len(loader.dataset)

    if epoch == 1 or epoch % 10 == 0:
        print(f"Epoch {epoch:3d} — Train MSE: {epoch_loss:.4f}")
````

## Leçon 4 — Exercice 2 : Solution Complète Annotée

✅ Solution complète et annotée de l'Exercice 2 du mock exam, avec des données synthétiques (remplace-les par le
vrai CSV le jour de l'examen : `data = pd.read_csv('data/exercise2.csv')`). Exécute-la et assure-toi de comprendre
chaque section marquée `── Tâche n ──`.

````python
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
from torch.utils.data import TensorDataset, DataLoader
from sklearn.model_selection import train_test_split

torch.manual_seed(42)

# ── Tâche 1: Charger les données ──────────────────────
# data = pd.read_csv('data/exercise2.csv')
# En exercice: X = toutes colonnes sauf target, y = colonne target

# ── Tâche 2: Qualité des données ──────────────────────
# number_of_nans = data.isna().sum().sum()
# Boxplot de chaque colonne pour détecter les outliers

# ── Tâche 3: Préparer les tenseurs ────────────────────
# X, y = data.iloc[:, features], data.iloc[:, target]
X_np = np.random.randn(500, 9).astype(np.float32)
y_np = (X_np[:, :3].sum(axis=1, keepdims=True)).astype(np.float32)

X_train, X_test, y_train, y_test = train_test_split(
    X_np, y_np, test_size=0.2, random_state=42)

batch_size = 256
train_loader = DataLoader(
    TensorDataset(torch.from_numpy(X_train), torch.from_numpy(y_train)),
    batch_size=batch_size, shuffle=True)
test_loader = DataLoader(
    TensorDataset(torch.from_numpy(X_test), torch.from_numpy(y_test)),
    batch_size=batch_size)

# ── Tâche 4: Définir le modèle ────────────────────────
class LinearReplicator(nn.Module):
    def __init__(self, in_features=9):
        super().__init__()
        self.lin = nn.Linear(in_features, 1, bias=True)  # poids + biais
    def forward(self, x):
        return self.lin(x)

model = LinearReplicator(in_features=X_train.shape[1])
criterion = nn.MSELoss()                           # perte MSE
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
n_epochs = 200

# ── Tâche 4: Boucle d'entraînement ────────────────────
for epoch in range(1, n_epochs + 1):
    model.train()
    epoch_loss = 0.0
    for xb, yb in train_loader:
        optimizer.zero_grad()     # 1. reset gradients
        y_pred = model(xb)        # 2. forward
        loss = criterion(y_pred, yb)  # 3. loss
        loss.backward()           # 4. backprop
        optimizer.step()          # 5. update
        epoch_loss += loss.item() * xb.size(0)
    epoch_loss /= len(train_loader.dataset)
    if epoch == 1 or epoch % 50 == 0:
        print(f"Epoch {epoch:3d} — Train MSE: {epoch_loss:.4f}")

# ── Tâche 5: Évaluation et scatter plot ───────────────
model.eval()
with torch.no_grad():
    all_preds, all_targets = [], []
    for xb, yb in test_loader:
        all_preds.append(model(xb))
        all_targets.append(yb)
    y_hat  = torch.vstack(all_preds)
    y_true = torch.vstack(all_targets)

mse_test = criterion(y_hat, y_true).item()
print(f"\nTest MSE: {mse_test:.4f}")

# Scatter: true vs predicted
plt.figure(figsize=(6, 6))
plt.scatter(y_true.numpy(), y_hat.numpy(), alpha=0.4, s=15)
plt.plot([y_true.min(), y_true.max()], [y_true.min(), y_true.max()], 'r--', lw=2)
plt.xlabel('True y'); plt.ylabel('Predicted y')
plt.title('Exercice 2: True vs Predicted')
plt.tight_layout(); plt.show()
````
